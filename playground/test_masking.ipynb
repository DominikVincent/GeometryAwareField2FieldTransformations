```python
from nerfstudio.model_components.nesf_components import *
import torch
```

```python
mc = MaskerConfig(
    mode="patch_fp",
    mask_ratio=0.5,
    num_patches=400
)


num_points = 1000
num_features = 48

masker: Masker = mc.setup(output_size=num_features)
```

```python
# generate data


points = torch.rand(1, num_points, 3)
outs = torch.arange(num_points).repeat(num_features, 1).transpose(0, 1).unsqueeze(0).float()

batch = {"points_xyz": points}

print("points_xyz: ", points.shape, "outs: ", outs.shape)
print(outs)
```

```python
outs, mask, ids_restore, batch = masker.mask(outs, batch)

print(outs, outs.shape)
print(outs.unique())
```

```python
outs, transform_batch = masker.unmask(outs, batch, ids_restore)

print(outs, outs.shape)
```
